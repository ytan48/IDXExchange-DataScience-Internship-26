```python
import sys
import joblib
import pandas as pd

sys.path.append("..")

from sklearn.linear_model import LinearRegression

from src.preprocessing import (
    FEATURES,
    TARGET,
    build_model_pipeline
)
```

```python
df = pd.read_csv(
    "../data/processed/crmls_modeling_splits.csv",
    dtype={"PostalCode": "string"},
    parse_dates=["CloseDate"]
)

train_df = df[df["split"] == "train"].copy()
validation_df = df[df["split"] == "validation"].copy()
test_df = df[df["split"] == "test"].copy()
```

```python
linear_pipeline = build_model_pipeline(
    model=LinearRegression(),
    scale_numeric=True
)

linear_pipeline.fit(
    train_df[FEATURES],
    train_df[TARGET]
)
```

```python
validation_predictions = linear_pipeline.predict(validation_df[FEATURES])
```

```python
r2_validation = linear_pipeline.score(validation_df[FEATURES], validation_df[TARGET])
print(f"R^2 (validation): {r2_validation:.4f}")
```
