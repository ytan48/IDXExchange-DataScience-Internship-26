```python
import sys
import pandas as pd
import numpy as np

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

sys.path.append("../src")

from preprocessing import (
    FEATURES,
    TARGET,
    build_model_pipeline
)
```

```python
df = pd.read_csv(
    "../data/processed/crmls_modeling_splits.csv",
    dtype={
        "PostalCode": "string"
    },
    low_memory=False
)

train_df = df[df["split"] == "train"].copy()
validation_df = df[df["split"] == "validation"].copy()
test_df = df[df["split"] == "test"].copy()

print("Train:", train_df.shape)
print("Validation:", validation_df.shape)
print("Test:", test_df.shape)
```

```python
X_train = train_df[FEATURES]
y_train = train_df[TARGET]

X_validation = validation_df[FEATURES]
y_validation = validation_df[TARGET]

X_test = test_df[FEATURES]
y_test = test_df[TARGET]
```

```python
tree_model = build_model_pipeline(
    DecisionTreeRegressor(
        max_depth=20,
        min_samples_leaf=10,
        random_state=42
    ),
    scale_numeric=False
)
```

```python
tree_model.fit(X_train, y_train)
```

```python
validation_pred = tree_model.predict(X_validation)

validation_r2 = r2_score(
    y_validation,
    validation_pred
)

print("Validation R²:", validation_r2)
```

```python
test_pred = tree_model.predict(X_test)

tree_test_r2 = r2_score(
    y_test,
    test_pred
)

print("Decision Tree Test R²:", tree_test_r2)
```

RT

```python
rf_model = build_model_pipeline(
    RandomForestRegressor(
        n_estimators=100,
        max_depth=20,
        min_samples_leaf=5,
        n_jobs=-1,
        random_state=42
    ),
    scale_numeric=False
)
```

```python
rf_model.fit(X_train, y_train)
```

```python
validation_pred = rf_model.predict(X_validation)

rf_validation_r2 = r2_score(
    y_validation,
    validation_pred
)

print("Random Forest Validation R²:", rf_validation_r2)
```

```python
test_pred = rf_model.predict(X_test)

rf_test_r2 = r2_score(
    y_test,
    test_pred
)

print("Random Forest Test R²:", rf_test_r2)
```

```python
comparison = pd.DataFrame({
    "Model": [
        "Decision Tree",
        "Random Forest",
        "Linear Regression"
    ],
    "Test R2": [
        tree_test_r2,
        rf_test_r2,
        round(0.8325690947501178, ndigits=6)
    ]
})

comparison
```
